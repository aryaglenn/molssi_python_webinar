# Tabular Data
## Processing tabular data using numpy

```python
import numpy
import os
```

```python
help(numpy.genfromtxt)
```

```python
distance_file = os.path.join("data", "distance_data_headers.csv")
distances = numpy.genfromtxt(fname=distance_file, delimiter=",", dtype="unicode")  # unicode means words or strings
print(distances)  # output is called a numpy array
```

```python
headers = distances[0]
print(headers)
```

```python
# slice distances to get the numerical values without header
data = distances[1:]
print(data)
```

```python
data = data.astype(numpy.float)  # cast whole array at once into numbers
print(data) # prints in scientific notation
```

```python
# to slice a 2D list
# array_name[row_index, column_index]

print(data[0, 1])
print(data[1, 0]) 
```

```python
small_data = data[0:10, 0:3]
```

```python
print(small_data)
```

```python
print(small_data[:, 1:])
```

```python
# data_average = numpy.mean(data_set)
```

```python
print(headers)
```

```python
thr4_atp = data[:, 1]
avg_thr4_atp = numpy.mean(thr4_atp)
print(avg_thr4_atp)
```

```python
# create a variable called num colums which is the number of columns
num_columns = len(headers)
```

```python
print(num_columns)
```

```python
for i in range(1, num_columns):
    #print(i)
    column = data[:, i]
    avg_col = numpy.mean(column)
    data_label = headers[i]
    print(F"{data_label}: {avg_col:.3f}")
```

# Plotting
## Plotting data using matplotlib

```python
import matplotlib.pyplot as plt  # define a shortcut for library so don't have to type the whole thing

%matplotlib inline
```

```python
plt.figure()  # create a new plot
plt.plot(data[:, 1])
```

```python
plt.figure()

sample_name = headers[1]  # what line corresponds to

plt.xlabel("Simulation Frame")  # x axis
plt.ylabel("Distance [angstrom]")  # y axis 
plt.plot(data[:, 1], label=sample_name)  # label is line
plt.title("Salad")  # title
plt.legend() # makes the label appear on plot

# to print or save figure
plt.savefig(F"{sample_name}.png", dpi = 300)  # dpi is resolution 
```

```python
plt.figure()

plt.xlabel("Simulation Frame")
plt.ylabel("Distance [angstrom]")
plt.plot(data[:, 1], label=headers[1])
plt.plot(data[:, 2], label=headers[2])
plt.legend()

plt.savefig("two_samples.png", dpi = 300)

        
```

```python
plt.figure() 

for col in range(1, num_columns):
    plt.plot(data[:, col], label = headers[col])
plt.xlabel("Simulation Frame")  # if not indented, won't be in for loop
plt.ylabel("Distance [angstrom]")
plt.legend()

plt.savefig("all_samples.png")
        
```

```python
# make a plot for each sample

# how to take min and max of data
data_min = numpy.min(data[:, 1:])  
data_max = numpy.max(data[:, 1:])

for col in range(1, num_columns):
    sample_name = headers[col]
    plt.figure() 
    plt.plot(data[:, col], label = sample_name)
    plt.ylim(data_min, data_max)  # y axis bounds
    
    plt.xlabel("Simulation Frame")  
    plt.ylabel("Distance [angstrom]")
    plt.legend()
    plt.savefig(F"{sample_name}.png") 
```

```python
plt.figure()
plt.plot(data[:, 0], data[:, 1], label = headers[1]) # (x, y)

plt.legend(loc="lower right")  # control location of label
```
